## 交叉熵损失（Cross Entropy）
衡量「模型预测的概率分布」和「真实标签的概率分布」之间的差距

假设单标签预测（即每个位置只有 1 个正确 token）场景。

loss = -ln(模型对正确 Token 的预测概率)

### 对数，描述的是什么，和熵有什么关系？

1. 对数，是在描述「一件事有多意外 / 有多少见 / 信息量有多大」。
    - 一件事概率越小（越罕见）
    - 它发生时带来的信息量越大
    - 这个信息量，用对数来算最自然
    - 数学上就是信息量 = -log (概率)，这就是对数最本质的意义：把概率 → 转成信息量
    
    - 举例
        - 太阳明天升起：概率 ≈ 1
            信息量 = 0（完全不意外）
        - 你买彩票中头奖：概率 ≈ 1/1000000
            信息量 = 非常大（超级意外）
2. 熵，就是信息量的期望值（熵 = 所有可能事件的「平均信息量」）。
    - 这个系统整体有多不确定、有多混乱、有多难预测
        - 熵高 = 很乱、很难猜、不确定性大
        - 熵低 = 很规律、很好猜、很确定

    - 举例
        - 扔均匀硬币：熵高（正反各 50%，很难猜）
        - 扔 “永远正面” 的硬币：熵 = 0（完全确定）

3. 为什么熵一定要用「对数」？
    - 熵 = 对每个事件：概率 × 该事件的信息量然后全部加起来求平均
    - 信息量 = -log (概率)
    - 所以，熵 = - Σ p (x) * log (p (x))



### 为什么用自然对数 ln ？不用其他的形式如 log10(x)？
1. 交叉熵的本质来自「信息论中的熵（Entropy）」，熵的定义天然使用自然对数 ln，PyTorch/TensorFlow 等框架中，F.cross_entropy 底层默认用自然对数计算
2. 深度学习的反向传播依赖「求导」，自然对数 ln(x) 的导数是 1/x（极简），而常用对数 log10(x) 的导数是 1/(x×ln(10))（多了一个常数项）；用 ln 能让梯度计算更简洁，减少不必要的计算量，提升训练效率。
3. 信息论中，“信息量” 的定义是 I(x) = -ln(P(x))（P (x) 是事件 x 发生的概率）
