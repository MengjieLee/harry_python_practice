# 线性结构回顾

## 什么是线性结构

- 线性结构是一种有序数据项的集合
- 每个数据项都有`唯一`的前趋和后继（除第一个和最后一个数据项外）
- 新的数据项加入到数据集合中时，只会加入到原有某个数据项之前或之后

具有这种性质的数据集，就称谓线性结构

如：高铁车厢


## 线性结构总有两端，在不同的情况下，两端的称呼也不同

- 有时候称为 左右
- 有时候称为 前后
- 有时候称为 顶底



### 两端的称呼不是关键，不同的线性结构关键在于数据项增减的方式，即操作方法的不同

- 有的结构只允许数据项从一端操作，而有的结构允许从两端操作
