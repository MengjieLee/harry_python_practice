# 线性表的案例引入

```python
# harry 小组任务汇报: 
# 1.是否到校了 in_school
# 2.是否递交作业 have_homework

six_group = [
    {
        "name": "A",
        "in_school": 1,
        "have_homework": 1
    },
    {
        "name": "B",
        "in_school": 0,
        "have_homework": 0
    }
]

def pan_duan_shifou_dao_xiao(six_group):
    # 循环 + 分支
    res = []
    for x in six_group:
        x_name = x["name"]
        x_school_status = "到了" if x["in_school"] == 1 else "没到"
        x_homework_status = "交了" if x["have_homework"] == 1 else "没交"

        x_stu = [x_name, x_school_status, x_homework_status]
        res.append(x_stu)
    return res

result = pan_duan_shifou_dao_xiao(six_group)
print(result) # [["A", "到了", "交了"],["B", "没到了", "没交"]]
```

## 一元多项式的运算


### Pn(x) = p0 + p1X + p2X^2 + ... + piXi + ... + pnX^n

```python
"""
线性表 P = (p0, p1, p2, ..., pn)
每一项的指数 i 隐藏在其系数 pi 的序号中
例如 P(x) = 10 + 5X - 4X^2 + 3X^3 + 2X^4
"""
p = [10, 5, -4, 3, 2] # 用数组表示，下标 i --> 指数, p[i] --> 系数的值

def generate_func():
    ans = []
    for coefficient, constant in enumerate(p):
        if coefficient == 0:
            ans.append(str(constant))
        else:
            ans.append(f'+{str(constant)}X^{coefficient}' if constant > 0 else f'{str(constant)}X^{coefficient}')
    return f"P(X) = {''.join(ans)}"

generate_func_instance = generate_func()
print(generate_func_instance)

```

# 总结

- 线性表中的`数据元素`的类型可以为`简单类型`（多项式运算案例,如整数 `int` 或小数 `float`），也可以为`复杂类型`（图书系统案例，如图书的序列号，书名，价格等等）
- 许多实际应用问题所涉及的基本操作有很大的相似性，不应为每个具体应用单独编写一个程序
- 从具体的应用中`抽象`出共性的`逻辑结构`和`基本操作`(抽象数据类型)，然后实现其`存储结构`和`基本操作`
